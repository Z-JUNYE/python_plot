### 初级绘制
先尝试用默认配置在同一张图上绘制正弦和余弦函数图像，然后逐步美化它。
第一步是取得正弦函数和余弦函数的值。

```python
import pandas as pd
import numpy as np
from pylab import *
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi*2, np.pi*2, 1024, endpoint=True)
#print(X)
```

```python
C,S = np.cos(X), np.sin(X)
print(C,S)
```

```python
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

### 以下是默认配置的具体内容

```python
figure(figsize = (8,6),dpi = 80) #创建一个8*6的图，并设置分辨率为80
subplot(1,1,1) #创建一个1*1的子图，接下来的图绘制在其中的第一块（也是唯一一块）
X = np.linspace(-np.pi*2, np.pi*2, 1024, endpoint=True)
C,S = np.cos(X), np.sin(X)

plot(X, C, color = 'blue', linewidth = 1.0, linestyle = '-') #绘制余弦函数，使用蓝色的、连续的、宽度为1像素的线条
plot(X, S, color = 'green', linewidth = 1.0, linestyle = '-') #绘制余弦函数，使用绿色的、连续的、宽度为1像素的线条

xlim=(-4.0, 4.0) # 设置横轴上下限
xticks(np.linspace(-4,4,9, endpoint=True)) #设置横轴标记

ylim=(-1.0,1.0) #设置竖轴上下限
yticks(np.linspace(-1,1,5,endpoint=True)) #设置竖轴标记

show() #在屏幕上显示
```

下面尝试改变线条颜色和粗细

```python
plot(X, C, color = 'red', linewidth = 2.5, linestyle = '-') #绘制余弦函数，使用红色的、连续的、宽度为2.5像素的线条
plot(X, S, color = 'yellow', linewidth = 2.5, linestyle = '--') #绘制余弦函数，使用黄色的、虚线的、宽度为2.5像素的线条
```

设置图片边界

```python
Y = C
xmin, xmax = X.min(), X.max()
ymin, ymax = Y.min(), Y.max()

dx, dy = (xmax-xmin)*0.2, (ymax-ymin)*0.2
xlim = (xmin-dx, xmax+dx)
ylim = (ymin-dy, ymax+dy)

show()
```
