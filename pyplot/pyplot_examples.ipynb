```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

`matplotlib.pyplot`是一个命令风格函数的集合，使`matplotlib`的机制更像MATLAB。  
每个绘图函数对图形进行一些更改。

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

当向`plot()`提供单个列表或者数组，那么`matplotlib`会假定它是一个y值序列，并自动生成x值。  
由于python范围从0开始，默认x向量具有与y相同的长度，单从0开始，所以上图中x值是[0,1,2,3]。
`plot()`是一个通用命令，并且可以接受任意数量的参数。

```python
a = [1,2,3,4]
b = [1,4,9,16]
plt.plot(a,b)
plt.show()
```

对于每个x和y参数对，还有一个可选的第三参数，它是指示图形颜色和线条类型的格式字符串。格式字符串的字母和符号来自于MATLAB，并且将颜色字符串与线型字符串连接在一起。
默认格式字符串是`b-`，它是一条蓝色实线。
如果要绘制红色圆圈，需要使用`ro`。

```python
plt.plot(a,b,'ro')
plt.axis([0,6,0,20])
plt.show()
```

如果`matplotlib`仅限于使用列表，它对数字处理是相当无用的。
一般来说，可以使用`numpy`数组。事实上，所有序列都在内部转换为`numpy`数组。
下面的实例展示了使用数组和不同格式字符串，在一条命令中绘制多个线条。

```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

MATLAB和pyplot具有当前图形和当前轴域的概念。
所有绘图命令适用于当前轴域。
